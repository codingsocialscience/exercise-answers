```r
## 3.15

candidate_comments <- 0 ## two gatherer variables
other_comments <- 0

data <- read.csv('twitter-conversation.txt', header = FALSE )

for( rn in 1:nrow( data ) ) {
    day <- data[ rn, 1]
    day <- gsub('Day', '', day )
    day <- as.integer( day )
    
    username <- data[rn, 2]

    if( day < 30 ) {
        if( username == 'hansjohansson' || username == 'evapersson' ) {
            candidate_comments <- candidate_comments + 1
        } else {
            other_comments <- other_comments + 1
        }
    } 
}
       
if( candidate_comments > 2 && other_comments > 1 ) {
    print("Discussion qualifies")
}
```
